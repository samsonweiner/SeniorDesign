```python
import pandas as pd
import os
```

```python
os.remove('dep2018.csv')
os.remove('dep2020.csv')
```

# Depression 2018

```python
file_in = open('dep2018.txt', 'r')
file_out = open('dep2018.csv', 'a')

for line in file_in:
    pieces = line.split()
    
    # Get State Name 
    state_pieces = pieces[1:len(pieces)-3]
    state=''
    for word in state_pieces:
        state += (' ' + word)
        
    # Depression Rate
    depression_rate = pieces[len(pieces)-3]
    
    out_str = state + ',' + depression_rate + '\n'
    
    file_out.write(out_str)
    
file_in.close()
file_out.close()
```

```python
# Test
dep2018 = pd.read_csv('dep2018.csv')
```

```python
# Reformat Depression Rate
for i in range(len(dep2018)):
    depRate = dep2018.loc[i, 'Depression_Rate']
    depRate = float(depRate[:len(depRate)-1])/100
    dep2018.loc[i, 'Depression_Rate'] = depRate
dep2018.head(15)
```

# Depression 2020

```python
file_in = open('dep2020.txt', 'r')
file_out = open('dep2020.csv', 'a')

for line in file_in:
    pieces = line.split()
    state_pieces = pieces[:len(pieces)-2]
    state = ''
    for word in state_pieces:
        state += (' ' + word)
    depRate = pieces[len(pieces)-1]
    
    out_str = state + ',' + depRate + '\n'
    
    file_out.write(out_str)
    
file_in.close()
file_out.close()
```

```python
dep2020 = pd.read_csv('dep2020.csv')
```

```python
dep2020.head()
```

```python
# Fix Depression Rate
# Reformat Depression Rate
for i in range(len(dep2020)):
    depRate = dep2020.loc[i, '2020_Growth']
    depRate = float(depRate[:len(depRate)-1])/100
    dep2020.loc[i, '2020_Growth'] = depRate
dep2020.head(15)
```

```python
# Gets rid of Washington DC
dep2020 = dep2020.drop([8]).reset_index().drop(['index'],axis=1)
```

```python
dep2020.head(10)
```

# Depression Rates 2020

```python
# Set Values 

# Loop Through each data frame
for i in range(len(dep2020)):
    for j in range(len(dep2018)):
        
        # When state names match
        if dep2020.loc[i, ' State'] == dep2018.loc[j, ' State']:
            
            # Initial Depression Rate
            old_rate = dep2018.loc[i, 'Depression_Rate']
            
            # Change In Rate 2020
            rate_change = dep2020.loc[i, '2020_Growth']
            
            # If rate is positive multiply by 1 + rate_change
            if rate_change >= 0:
                new_rate = old_rate * (1 + rate_change) 
            
            # If rate is negative multiply by 1 - rate_change
            else:
                new_rate = old_rate * (1 - abs(rate_change))
                
            # Set value accordingly
            # Values are changed in place of dep2018
            dep2018.loc[i, 'Depression_Rate'] = new_rate
```

```python
# Simply copy 2018 to the new 2020 Frame
dep2020 = dep2018.copy()
```

```python
# Write this new depression rate to a csv
dep2020.to_csv('dep2020_Actual.csv')
```
